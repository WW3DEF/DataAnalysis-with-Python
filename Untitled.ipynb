```python
import csv
f = open('busan_age.csv', encoding='utf8')
data = csv.reader(f)

header = next(data)
result = []

name = input('인구구조를 보고 싶은 지역을 입력하세요! : ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            i = i.replace(',', '')
            result.append(int(i))
            
print(result)      

import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic') # 한글처리
plt.rcParams['axes.unicode_minus'] = False # - 부호 나오기
plt.title(name)
# plt.plot(result, 'ro-', label='인구수')

plt.bar(range(101), result, label='인구수')

# plt.legend()
plt.show()
```
