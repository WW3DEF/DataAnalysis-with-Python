```python
import csv

f = open('busan_age.csv', encoding='utf8')
data = csv.reader(f)

header = next(data)
result = []
result2 = []
result3 = []
result4 = []

name = input('인구구조를 보고 싶은 지역을 입력하세요! : ')
name2 = input('인구구조를 보고 싶은 2번째 지역을 입력하세요! : ')
name3 = input('인구구조를 보고 싶은 3번째 지역을 입력하세요! : ')
name4 = input('인구구조를 보고 싶은 4번째 지역을 입력하세요! : ')

for row in data:
    if name in row[0]: # 1번째 입력한 값과 일치하는 행정구역
        for i in row[3:]: # 0세부터 끝(100세)까지
            i = i.replace(',', '') # 컴마(,)를 공백으로 변경
            result.append(int(i))
    if name2 in row[0]: # 2번째 입력한 값과 일치하는 행정구역
        for i in row[3:]:
            i = i.replace(',', '')
            result2.append(int(i))
    if name3 in row[0]: # 3번째 입력한 값과 일치하는 행정구역
        for i in row[3:]:
            i = i.replace(',', '')
            result3.append(int(i))
    if name4 in row[0]: # 4번째 입력한 값과 일치하는 행정구역
        for i in row[3:]:
            i = i.replace(',', '')
            result4.append(int(i))
print(result)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot') # 그래프 스타일 격자무늬(ggplot) 지정
plt.rc('font', family ='Malgun Gothic') #폰트 '맑은 고딕' 지정
plt.rcParams['axes.unicode_minus'] = False # - 부호 나오기
plt.title('부산의 구별 인구분포') # 그래프 제목
plt.plot(result, 'ro-', label=name) # 빨강 원마커 실선
plt.plot(result2, 'bx-', label=name2) # 파랑 x마커 실선
plt.plot(result3, 'g^-', label=name3) # 초록 삼각마커 실선
plt.plot(result4, 'ys-', label=name4) # 노랑 사각마커 실선
plt.legend()
plt.show()
```
