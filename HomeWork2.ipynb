```python
import csv
import math

# 데이터 불러오기
f = open('age-mf.csv', encoding='utf-8')
data = csv.reader(f)
header = next(data)

a = input("원하는 지역을 입력해주세요 : ") # 입력받기
resultM = [] # 남성 데이터 리스트
resultF = [] # 여성 데이터 리스트
size = [] # 마커 사이즈 지정할 리스트
for row in data : # 데이터 전부 출력하기
    if a in row[0] : # 입력받은 값과 일치하는 행이면
        for i in range(101): # 0 ~ 100까지
            resultM.append(int(row[i+3].replace(',', ''))) # 0세부터 resultM 리스트에 데이터 삽입
            resultF.append(int(row[i+106].replace(',', ''))) # 0세부터 resultF 리스트에 데이터 삽입
            size.append(math.sqrt(resultM[i]+resultF[i])) # size 리스트에 resultM + resultF 값을 제곱근 함수(math.sqrt)한 결과 삽입
        break
```

```python
import matplotlib.pyplot as plt
plt.title(f'{a}의 성별 인구 그래프') # 타이틀 지정
plt.rc('font',family='Malgun Gothic') #폰트 '맑은 고딕' 지정
plt.rcParams['axes.unicode_minus'] = False # - 부호 나오기
plt.scatter(resultM, resultF, s=size, c=range(101), cmap='jet') # 산점도 넣기 ( 남성 데이터, 여성 데이터, 마커 크기를 size변수로 지정, 100까지 컬러지정, 컬러맵 스타일 'jet' )
plt.colorbar() # 컬러바 생성
plt.plot(range(max(resultM)),range(max(resultM)), 'g') # 추세선 긋기
plt.style.use('ggplot') # 차트 스타일 지정
plt.xlabel('남성 인구수') # x축 레이블
plt.ylabel('여성 인구수') # y축 레이블
plt.show() # 차트 보이기
```
